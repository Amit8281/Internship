## 1. Scrape the details of most viewed videos on YouTube from Wikipedia.
### Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
### You need to find following details:
- A) Rank
- B) Name
- C) Artist
- D) Upload date
- E) Views

```python
#Importing the libraries

import pandas as pd
import re
import selenium
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import time
from selenium.common.exceptions import ElementNotInteractableException
from selenium.common.exceptions import NoSuchElementException

from selenium.webdriver.chrome.options import Options     #importing Options
chrome_options = Options()
chrome_options.add_argument('--incognito')

import warnings
warnings.filterwarnings('ignore')
```

```python
chrome_options.add_argument('--incognito')
driver = webdriver.Chrome('chromedriver.exe')

driver.maximize_window() 
time.sleep(1)
```

```python
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
```

```python
#Extracting the details from the website.

rank = []
name = []
artist = []
date = []
views = []

#Extracting the rank details
try:
    rank_tags = driver.find_elements(By.XPATH,"//table[@class = 'wikitable sortable jquery-tablesorter']/tbody/tr/td[1]")
    for ra in rank_tags:
        rank.append(ra.text)
    rank = rank[:30]
except NoSuchElementException:
    Company.append('NA')


#Extracting the name details
try:
    name_tags = driver.find_elements(By.XPATH,"//table[@class = 'wikitable sortable jquery-tablesorter']/tbody/tr/td[2]")
    for na in name_tags:
        name.append(na.text[:-3].replace('[',''))

    name = name[:30]
except NoSuchElementException:
    Company.append('NA')


#Extracting the artist details
try:
    artist_tags = driver.find_elements(By.XPATH,"//table[@class = 'wikitable sortable jquery-tablesorter']/tbody/tr/td[3]")
    for art in artist_tags:
        artist.append(art.text)
    artist = artist[:30]
except NoSuchElementException:
    Company.append('NA')


#Extracting the upload date details
try:
    date_tags = driver.find_elements(By.XPATH,"//table[@class = 'wikitable sortable jquery-tablesorter']/tbody/tr/td[5]")
    for dt in date_tags:
        date.append(dt.text)
    date = date[:30]
except NoSuchElementException:
    Company.append('NA')


#Extracting the Views details
try:
    view_tags = driver.find_elements(By.XPATH,"//table[@class = 'wikitable sortable jquery-tablesorter']/tbody/tr/td[4]")
    for vi in view_tags:
        views.append(vi.text)
    views = views[:30]
except NoSuchElementException:
    Company.append('NA')


#Creating the dateframe to display the scraped content
youtube_top = pd.DataFrame({})
youtube_top['Rank'] = rank
youtube_top['Video Name'] = name
youtube_top['Artist/Uploader'] = artist
youtube_top['Upload Date'] = date
youtube_top['Views(In Billions)'] = views

youtube_top
```

```python
driver = webdriver.Chrome('chromedriver.exe')
driver.maximize_window()
```

```python
driver.get(" https://www.naukri.com/hr-recruiters-consultants")
time.sleep(1)
```

```python
# Finding the search bar and entering the designation to search for
designation = driver.find_element(By.CLASS_NAME, "sugInp")
designation.send_keys('Data Science')
```

```python
# Clicking the search button
search = driver.find_element(By.XPATH, "/html/body/div[2]/div[2]/div[1]/div[1]/form/div[1]/button")
search.click()
```

```python
# Creating empty lists to store the scraped data
Name =[]
Designation =[]
Company =[]
Skill =[]
Location =[]

# Scraping Name
try:
    name = driver.find_elements(By.XPATH, '//a[@class="ellipsis"]/span')
    for i in name:
        Name.append(i.text)
except NoSuchElementException:
    Name.append('NA')

# Scraping Designation
try:
    designation = driver.find_elements(By.XPATH, '//p[@class="highlightable"]/span[1]')
    for i in designation:
        Designation.append(i.text)
except NoSuchElementException:
    Designation.append('NA')

# Scraping Company
try:
    company = driver.find_elements(By.XPATH, '//a[@class="ellipsis"][2]')
    for i in company:
        Company.append(i.text)
except NoSuchElementException:
    Company.append('NA')

# Scraping skills they hire for
try:
    skill = driver.find_elements(By.XPATH, '//div[@class="hireSec highlightable"]')
    for i in skill:
        Skill.append(i.text)
except NoSuchElementException:
    Skill.append('NA')

# Scraping Location
try:
    location = driver.find_elements(By.XPATH, '//p[@class="highlightable"]/span[2]/small')
    if location:
        for i in location:
            Location.append(i.text)
    else:
        Location.append('NA')
except NoSuchElementException:
    Location.append('NA')

# Creating a pandas dataframe to store the scraped data
Naukri=pd.DataFrame({})
Naukri['Name']=Name[:49]
Naukri['Designation']=Designation[:49]
Naukri['Company']=Company[:49]
Naukri['Skills I Hire for']=Skill[:49]
Naukri['Location']=Location[:49]
```
